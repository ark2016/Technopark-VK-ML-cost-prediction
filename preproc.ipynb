# нормализация данных и обучение

```python
categorial_features = ('street_id', 'balcon', 'rooms')
numeric_features = ('date', 'g_lift', 'build_tech',
                    'floor', 'area', 'metro_dist',
                    'kw1', 'kw2', 'kw3',
                    'kw4', 'kw5', 'kw6',
                    'kw7', 'kw8', 'kw9',
                    'kw10', 'kw11', 'kw12',
                    'kw13')
```

```python
from sklearn.preprocessing import OneHotEncoder, MaxAbsScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from catboost import CatBoostRegressor
```

```python
num_pipeline = Pipeline([
    ('scaler', MaxAbsScaler())
])
```

```python
pipe = Pipeline([
    ('preproc', ColumnTransformer([
            ('numeric', num_pipeline, numeric_features),
            ('categorial', OneHotEncoder(), categorial_features)
        ]
    )),
    ('grad_boost', CatBoostRegressor(iterations=1000,
                           task_type="GPU",
                           devices='0'))
])
```

```python
X = train.drop(columns=["price"])
X
```

```python
y = train["price"]
y
```

```python
pipe.fit(X, y)
```

```python
ans = pd.read_csv("sample_submission.csv")
ans
```

```python
res = pipe.predict(test)
res
```

```python
ans.to_csv("preproc_first_try.csv", index=False)
```
